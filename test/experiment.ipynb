```python
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

RANDOM_STATE = 5

b = datasets.load_breast_cancer()
(X_all, y_all) = b.data, b.target
X_train, X_test, y_train, y_test = train_test_split(X_all, y_all, test_size=0.25, random_state=RANDOM_STATE)
```

```python
n = KNeighborsClassifier(n_neighbors=5)
n.fit(X_train, y_train)
```

```python
print("SCIKIT train score", n.score(X_train, y_train))
print("SCIKIT test score", n.score(X_test, y_test))
```

```python
import numpy as np

NUMBER_NEIGHBORS = 5

def group_prediction(y_hat):
  if y_hat >= 0.5:
    return 1
  else:
    return 0

def euclidean_distance(xi, xj):
  return (np.linalg.norm(xi - xj))

for xi in X_train:
  dist = np.apply_along_axis(euclidean_distance, 1, X_train, xi)
  sorted_dist_indexes = np.argsort(dist)
  neighbors = sorted_dist_indexes[1:NUMBER_NEIGHBORS+1]
  mean_label = y_train[neighbors].mean()
  print(neighbors)
  print(y_train[neighbors].mean())
  print(group_prediction(mean_label))
```
